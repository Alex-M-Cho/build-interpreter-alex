## Upgrade 1: Parse numbers

```python
def tokenize(text):
    tokens = []
    current = ""
    for char in text:
        if char.isdigit():
            current += char
        else:
            if current:
                tokens.append(current)
                current = ""
        
            tokens.append(char)
    
    if current:
        tokens.append(current)
        
    return tokens
```

```python
tokenize("1+22")
```

```python
tokenize("23*10")
```

# Upgrade 2: Label the token

```python
def tokenize(text):
    tokens = []
    current = ""
    for char in text:
        if char.isdigit():
            current += char
        else:
            if current:
                tokens.append(("Number", current))
                current = ""
        
            tokens.append(("Operator", char))
    
    if current:
        tokens.append(("Number", current))
        
    return tokens
```

```python
tokenize("23*10")
```
