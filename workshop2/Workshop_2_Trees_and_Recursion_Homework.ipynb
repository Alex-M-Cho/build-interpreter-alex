# Part 1 — Representing Trees in Python

A tree is made of **nodes**. Each node stores:

- a value
- zero or more children

Example:

```text
Documents
├── Photos
└── School
```

`Documents` is the root. `Photos` and `School` are children.

````python
@dataclass
class Node:
    value: Any
    children: List["Node"] = field(default_factory=list)

    def add_child(self, child: "Node") -> None:
        self.children.append(child)
````

# Part 2 — Recursion Warmup

A recursive function needs:

1. A **base case**
2. A **recursive case**

The base case stops the recursion. The recursive case solves a smaller version of the same problem.

## Exercise 4 — Recursive Factorial

Complete the recursive factorial function.

```text
factorial(5) = 5 × factorial(4)
factorial(4) = 4 × factorial(3)
...
factorial(1) = 1
```

````python
def factorial(n: int) -> int:
    # TODO: base case
    if n == 0:
        return 1

    # TODO: recursive case
    return n * factorial(n - 1)

# Expected: 120
factorial(5)
````

# Part 3 — Recursive Functions on Trees

Trees are naturally recursive because each child is itself the root of a smaller tree.

```text
tree = current node + child subtrees
```

## Exercise 6 — Count All Nodes

Hint:

```text
count_nodes(node) = 1 + count_nodes(each child)
```

````python
def count_nodes(node: Node) -> int:
    total = 1  # count the current node

    for child in node.children:
        total += count_nodes(child)

    return total

# Expected: 7
count_nodes(documents)
````

## Exercise 7 — Count Leaves

A leaf is a node with no children.

````python
def count_leaves(node: Node) -> int:
    # TODO: base case
    if len(node.children) == 0:
        return 1

    total = 0

    for child in node.children:
        total += count_leaves(child)

    return total

# Expected: 4
count_leaves(documents)
````

## Exercise 8 — Compute Tree Height

In this homework, define height as the number of edges on the longest path from a node to a leaf. A leaf has height `0`.

Example:

```text
A
├── B
└── C
    └── D
```

The height of `A` is `2`.

````python
def height(node: Node) -> int:
    if len(node.children) == 0:
        return 0

    child_heights = []

    for child in node.children:
        child_heights.append(height(child))

    return 1 + max(child_heights)

# Expected: 2
height(documents)
````

# Part 4 — Tree Traversal

Traversal means visiting every node.

We will implement **preorder traversal**:

```text
visit current node
then visit children
```

## Exercise 9 — Preorder Traversal

Return a list of values in preorder.

````python
def preorder(node: Node) -> list:
    result = [node.value]  # current node first

    for child in node.children:
        result.extend(preorder(child))

    return result

# Expected:
# ['Documents', 'Photos', 'beach.jpg', 'dog.png', 'School', 'CS101', 'Notes']
preorder(documents)
````

# Part 5 — Expression Trees

Expression trees represent arithmetic expressions.

Example:

```text
1 + 2 * 3
```

becomes:

```text
      +
     / \
    1   *
       / \
      2   3
```

For expression trees:

- number nodes are leaves
- operator nodes have a left child and a right child

````python
@dataclass
class ExprNode:
    value: Any
    left: Optional["ExprNode"] = None
    right: Optional["ExprNode"] = None

    @property
    def is_number(self) -> bool:
        return isinstance(self.value, (int, float))

    @property
    def is_operator(self) -> bool:
        return self.value in ["+", "-", "*", "/", "**"]
````

## Exercise 11 — Build an Expression Tree

## Exercise 12 — Evaluate an Expression Tree

Complete the recursive evaluator.

Strategy:

```text
If node is a number:
    return the number

Otherwise:
    evaluate left child
    evaluate right child
    combine using the operator
```

````python
def evaluate(node: ExprNode) -> float:
    if node.is_number:
        return node.value

    left = evaluate(node.left)
    right = evaluate(node.right)

    if node.value == "+":
        return left + right
    elif node.value == "-":
        return left - right
    elif node.value == "*":
        return left * right
    elif node.value == "/":
        return left / right
    elif node.value == "**":
        return left ** right
    else:
        raise ValueError(f"Unknown operator: {node.value}")

# Expected: 7
evaluate(expr)
````

## Exercise 14 — Pretty Print Expression Trees

Write a function that turns an expression tree back into a string.

Example:

```python
pretty(expr)
```

should return:

```text
(1 + (2 * 3))
```

````python
def pretty(node: ExprNode) -> str:
    if node.is_number:
        return str(node.value)

    left = pretty(node.left)
    right = pretty(node.right)

    return "(" + left + " " + node.value + " " + right + ")"

pretty(expr)
````

Write a function that returns all operators in an expression tree.

Example:


should return:


````python
def operators(node: ExprNode) -> list:
    if node.is_number:
        return []

    result = [node.value]

    result.extend(operators(node.left))
    result.extend(operators(node.right))

    return result

operators(expr)
````

Write a function that checks if an expression tree is valid.

Rules:

- number nodes should not have children

- unknown operators are invalid

````python
def is_valid_expr_tree(node: ExprNode) -> bool:
    # TODO: implement from scratch
    if node is None:
        return False
    if node.is_number:
        return node.left is None and node.right is None
    if node.is_operator:
        return is_valid_expr_tree(node.left) and is_valid_expr_tree(node.right)
    return False

# Try:
is_valid_expr_tree(expr)
````
