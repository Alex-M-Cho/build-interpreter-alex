```python
from dataclasses import dataclass
from typing import Any


@dataclass
class ExprNode:
    value: Any
    left: "ExprNode | None" = None
    right: "ExprNode | None" = None

    @property
    def is_number(self):
        return isinstance(self.value, (int, float))


# Expression:
#
#       +
#      / \
#     1   *
#        / \
#       2   3
#
# Represents:
# 1 + (2 * 3)

root = ExprNode(
    "+",
    ExprNode(1),
    ExprNode(
        "*",
        ExprNode(2),
        ExprNode(3)
    )
)


def evaluate(node):
    """
    Evaluate an expression tree.
    """
    if not node.left and not node.right:  # base case
        return node.value
    
    left = evaluate(node.left)   # recursive case
    right = evaluate(node.right)

    if node.value == "+":
        return left + right
    if node.value == "*":
        return left * right
        
print(evaluate(root))
```

```python
evaluate(ExprNode(3)) # return 7 
```

```python
def factorial(n):
    if n==1: return 1
    return n * factorial(n-1)
```
